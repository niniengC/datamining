# Getting to Know Yor Data


## Data Objects and Attribute Types
Data set terdiri dari objek data. Objek data mewakili entitas—dalam database penjualan, objek tersebut mungkin berupa pelanggan, barang toko, dan penjualan; dalam database medis, objeknya mungkin pasien; dalam database universitas, objeknya mungkin mahasiswa, profesor, dan kursus. Objek data biasanya dijelaskan oleh atribut. Objek data juga dapat disebut sebagai sampel, contoh, instance, data points, atau objects. Jika objek data disimpan dalam database, mereka adalah data tuple. Artinya, baris database sesuai dengan objek data, dan kolom sesuai dengan atribut. Di bagian ini, kita mendefinisikan atribut dan melihat berbagai tipe atribut.

### What Is an Attribute?
Atribut adalah data field, yang mewakili karakteristik atau fitur dari objek data. Atribut kata benda, dimensi, fitur, dan variabel sering digunakan secara bergantian dalam literatur. Istilah dimensi biasanya digunakan dalam data warehousing. Literatur pembelajaran mesin cenderung menggunakan istilah fitur, sedangkan ahli statistik lebih memilih istilah variabel. Data mining dan profesional database biasanya menggunakan istilah atribut, dan kami juga melakukannya di sini. Atribut yang menjelaskan objek pelanggan dapat mencakup, misalnya, ID pelanggan, nama, dan alamat. Nilai yang diamati untuk atribut tertentu dikenal sebagai observasi. Seperangkat atribut yang digunakan untuk menggambarkan objek tertentu disebut vektor atribut (atau vektor fitur). Distribusi data yang melibatkan satu atribut (atau variabel) disebut univariat. Distribusi bivariat melibatkan dua atribut, dan seterusnya. Jenis atribut ditentukan oleh kumpulan nilai yang mungkin—nominal, biner, ordinal, atau numerik—atribut dapat miliki. 

#### Atribut Nominal
Nominal berarti “berkaitan dengan nama”. Nilai atribut nominal adalah simbol atau nama benda. Setiap nilai mewakili beberapa jenis kategori, kode, atau keadaan, dan atribut nominal juga disebut sebagai kategorikal. Nilai-nilai tidak memiliki urutan yang berarti. Dalam ilmu komputer, nilai-nilai juga dikenal sebagai enumerasi.

**Contoh atribut Nominal :**
Misalkan warna rambut dan status perkawinan adalah dua atribut
menggambarkan objek orang. Dalam penerapannya, nilai yang mungkin untuk warna rambut adalah hitam, coklat, pirang, merah, pirang, abu-abu, dan putih. Atribut status perkawinan dapat mengambil nilai lajang, menikah, bercerai, dan janda. Baik warna rambut dan status perkawinan adalah atribut nominal. Contoh atribut nominal lainnya adalah pekerjaan, dengan nilai guru, dokter gigi, programer, petani, dan sebagainya. 

Meskipun dikatakan bahwa nilai atribut nominal adalah simbol atau "nama benda", simbol atau "nama" tersebut dapat direpresentasikan dengan angka. Dengan warna rambut misalnya, kita bisa memberi kode 0 untuk hitam, 1 untuk cokelat, dan seterusnya. Contoh lain adalah ID pelanggan, dengan kemungkinan nilai yang semuanya numerik. Namun, dalam kasus seperti itu, angka tersebut tidak dimaksudkan untuk digunakan secara kuantitatif. Artinya, operasi matematika pada nilai atribut nominal tidak bermakna. Tidak masuk akal untuk mengurangi satu nomor ID pelanggan dari yang lain, tidak seperti, katakanlah, mengurangkan nilai usia dari yang lain (di mana usia adalah atribut numerik). Meskipun atribut nominal mungkin memiliki bilangan bulat sebagai nilai, itu tidak dianggap sebagai atribut numerik karena bilangan bulat tidak dimaksudkan untuk digunakan secara kuantitatif. Kami akan membahas lebih lanjut tentang atribut numerik di Bagian Atribut Numerik. 

Karena nilai atribut nominal tidak memiliki urutan yang berarti tentangnya dan tidak kuantitatif, tidak masuk akal untuk menemukan nilai rata-rata (rata-rata) atau nilai median (tengah) untuk atribut semacam itu, mengingat sekumpulan objek. Namun, satu hal yang menarik adalah nilai atribut yang paling sering muncul. Nilai ini dikenal sebagai modus, merupakan salah satu ukuran tendensi sentral.
