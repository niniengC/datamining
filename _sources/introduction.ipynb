# INTRODUCTION

*   Kebutuhan adalah *mother of invention*. Dengan pertumbuhan data yang meningkat di setiap aplikasi, *Data Mining* memenuhi kebutuhan mendesak akan analisis data yang efektif, skalabel, dan fleksibel di masyarakat kita. *Data Mining* dapat dianggap sebagai evolusi alami teknologi informasi dan pertemuan beberapa disiplin ilmu dan domain aplikasi terkait.

*   *Data mining* adalah proses menemukan pola yang menarik dari sejumlah besar data. Sebagai proses penemuan pengetahuan, biasanya melibatkan *data cleaning, data integration, data selection, data transformation, pattern discovery, pattern evaluation,* dan *knowledge presentation*.

*   Suatu pola menarik jika valid pada data uji dengan tingkat kepastian tertentu, baru, berpotensi berguna (misalnya, dapat ditindaklanjuti atau memvalidasi firasat yang membuat penasaran pengguna), dan mudah dipahami oleh manusia. Pola yang menarik mewakili **pengetahuan**. Ukuran **ketertarikan pola**, baik *objektif* maupun *subjektif*, dapat digunakan untuk memandu proses *discovery*.

*   Kami menyajikan **multidimensional view** dari data mining. Dimensi utama adalah data, pengetahuan, teknologi, dan aplikasi.

*   Data mining dapat dilakukan pada semua jenis data selama data tersebut bermakna untuk aplikasi target, seperti data database, data warehouse, data transaksional, dan tipe data lanjutan.

*   Data mining dapat dilakukan pada semua jenis data selama data tersebut bermakna untuk aplikasi target, seperti data database, data gudang data, data transaksional, dan tipe data lanjutan. Tipe data tingkat lanjut termasuk time-related atau sequence data, data streams, spatial dan spatiotemporal data, text and multimedia data, graph dan networked data, dan Web data.

*   Warehouse adalah repository untuk penyimpanan data jangka panjang dari berbagai sumber, diatur sedemikian rupa untuk memfasilitasi pengambilan keputusan manajemen. Data disimpan di bawah skema terpadu dan biasanya diringkas. Sistem warehouse menyediakan kemampuan analisis data multidimensi, yang secara kolektif disebut sebagai **online analytical processing**.

*   **Multidimensional data mining** (juga disebut exploratory multidimensional data mining) mengintegrasikan teknik penambangan data inti dengan analisis multidimensi berbasis OLAP. Ini mencari pola yang menarik di antara beberapa kombinasi dimensi (atribut) pada berbagai tingkat abstraksi, sehingga menjelajahi ruang data multidimensi.

*   Fungsionalitas data mining digunakan untuk menentukan jenis pola atau pengetahuan yang ditemukan dalam tugas data mining. Fungsinya meliputi *characterization* dan discrimination; *the mining of frequent patterns, associations*, dan *correlations; classification* dan *regression; cluster analysis*; dan *outlier detection*. Karena jenis data baru, aplikasi baru, dan tuntutan analisis baru terus bermunculan, tidak diragukan lagi kita akan melihat semakin banyak tugas penambangan data baru di masa mendatang.

*   Penambangan data, sebagai domain yang sangat didorong oleh aplikasi, telah menggabungkan teknologi dari banyak domain lainnya. Ini termasuk *statistik, machine learning, database* dan *data warehouse systems*, dan *information retrieval*. Sifat interdisipliner penelitian dan pengembangan data mining memberikan kontribusi yang signifikan terhadap keberhasilan data mining dan aplikasi yang luas.

*   Data mining memiliki banyak aplikasi yang sukses, seperti *business intelligence, Web search, bioinformatics, health informatics, finance, digital libraries*, dan *digital governments*.

*   Ada banyak masalah yang menantang dalam penelitian data mining. Area termasuk metodologi penambangan, interaksi pengguna, efisiensi dan skalabilitas, dan berurusan dengan beragam tipe data. Penelitian penambangan data telah berdampak kuat pada masyarakat dan akan terus berlanjut di masa depan.
