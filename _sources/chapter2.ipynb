# Getting to Know Your Data

## Data Objek dan Tipe Atribut
Kumpulan data terdiri dari objek data. Objek data mewakili entitas—dalam database penjualan, objek tersebut mungkin berupa pelanggan, barang toko, dan penjualan; dalam database medis, objeknya mungkin pasien; dalam database universitas, objeknya mungkin mahasiswa, profesor, dan kursus. Objek data biasanya dijelaskan oleh atribut. Objek data juga dapat disebut sebagai sampel, contoh, contoh, poin data, atau objek. Jika objek data disimpan dalam database, itu adalah tupel data. Artinya, baris database sesuai dengan objek data, dan kolom sesuai dengan atribut. 

### Apa itu Atribut?
Atribut adalah bidang data, yang mewakili karakteristik atau fitur dari objek data. Atribut kata benda, dimensi, fitur, dan variabel sering digunakan secara bergantian dalam literatur. Istilah dimensi umumnya digunakan dalam *data werehousing*. Literatur pembelajaran mesin cenderung menggunakan istilah fitur, sedangkan ahli statistik lebih menyukai istilah variabel. Penambangan data dan profesional basis data biasanya menggunakan istilah atribut, dan kami juga melakukannya di sini. Atribut yang mendeskripsikan objek pelanggan dapat mencakup, misalnya, ID pelanggan, nama, dan alamat. Nilai yang diamati untuk atribut tertentu dikenal sebagai pengamatan. Seperangkat atribut yang digunakan untuk mendeskripsikan objek tertentu disebut vektor atribut (atau vektor fitur). Distribusi data yang melibatkan satu atribut (atau variabel) disebut univariat. Distribusi bivariat melibatkan dua atribut, dan seterusnya.
Jenis atribut ditentukan oleh kumpulan nilai yang mungkin—nominal, biner, ordinal, atau numerik—yang dapat dimiliki atribut tersebut.

### Tipe Atribut

#### Atribut Nominal
Nominal berarti "berkaitan dengan nama." Nilai atribut nominal adalah simbol atau nama benda. Setiap nilai mewakili beberapa jenis kategori, kode, atau keadaan, sehingga atribut nominal juga disebut sebagai kategori. Nilai tidak memiliki urutan yang berarti. Dalam ilmu komputer, nilai juga dikenal sebagai pencacahan.

**Contoh atribut Nominal :**

Misalkan warna rambut dan status perkawinan adalah dua atribut yang menggambarkan objek seseorang. Dalam aplikasi kita, nilai yang mungkin untuk warna rambut adalah hitam, coklat, pirang, merah, pirang, abu-abu, dan putih. Atribut status perkawinan dapat mengambil nilai lajang, menikah, bercerai, dan janda. Warna rambut dan status perkawinan adalah atribut nominal. Contoh atribut nominal lainnya adalah pekerjaan, dengan nilai guru, dokter gigi, programer, petani, dan sebagainya.

Meskipun dikatakan bahwa nilai atribut nominal adalah simbol atau "nama benda", simbol atau "nama" tersebut dapat direpresentasikan dengan angka. Dengan warna rambut misalnya, kita bisa memberi kode 0 untuk hitam, 1 untuk cokelat, dan seterusnya. Contoh lainnya adalah customor ID, dengan kemungkinan nilai yang semuanya numerik. Namun, dalam kasus seperti itu, angka tersebut tidak dimaksudkan untuk digunakan secara kuantitatif. Artinya, operasi matematis pada nilai atribut nominal tidak bermakna. Tidak masuk akal untuk mengurangi satu nomor ID pelanggan dari yang lain, tidak seperti, katakanlah, mengurangkan nilai usia dari yang lain (di mana usia adalah atribut numerik). Meskipun atribut nominal mungkin memiliki bilangan bulat sebagai nilai, itu tidak dianggap sebagai atribut numerik karena bilangan bulat tidak dimaksudkan untuk digunakan secara kuantitatif.
Karena nilai atribut nominal tidak memiliki urutan yang berarti tentangnya dan tidak kuantitatif, tidak masuk akal untuk menemukan nilai rata-rata (rata-rata) atau nilai median (tengah) untuk atribut seperti itu, mengingat sekumpulan objek. Namun, satu hal yang menarik adalah nilai atribut yang paling sering muncul. Nilai ini dikenal sebagai modus, merupakan salah satu ukuran tendensi sentral.

#### Atribut Biner
Atribut biner adalah atribut nominal dengan hanya dua kategori atau status: 0 atau 1, di mana 0 biasanya berarti atribut tersebut tidak ada, dan 1 berarti ada. Atribut biner disebut sebagai Boolean jika kedua status sesuai dengan benar dan salah.

**Contoh Atribut biner :**

Mengingat atribut perokok menggambarkan objek pasien, 1 menunjukkan bahwa pasien merokok, sedangkan 0 menunjukkan bahwa pasien tidak merokok. Demikian pula, misalkan pasien menjalani tes medis yang memiliki dua kemungkinan hasil. Tes medis atribut bersifat biner, dimana nilai 1 berarti hasil tes untuk pasien positif, sedangkan 0 berarti hasilnya negatif.
Atribut biner adalah simetris jika kedua statusnya sama-sama berharga dan memiliki bobot yang sama; yaitu, tidak ada preferensi hasil mana yang harus diberi kode 0 atau 1. Salah satu contohnya adalah atribut jenis kelamin yang memiliki status pria dan wanita.
Atribut biner adalah asimetris jika hasil dari keadaan tidak sama pentingnya, seperti hasil positif dan negatif dari tes medis untuk HIV. Dengan konvensi, kami memberi kode hasil yang paling penting, yang biasanya paling langka, dengan 1 (misalnya, HIV positif) dan yang lainnya dengan 0 (misalnya, HIV negatif).

#### Atribut Ordinal
Atribut ordinal adalah atribut dengan kemungkinan nilai yang memiliki urutan atau peringkat yang bermakna di antara mereka, tetapi besarnya antara nilai-nilai yang berurutan tidak diketahui.

**Contoh atribut ordinal :**

Misalkan ukuran minuman sesuai dengan ukuran minuman yang tersedia di restoran cepat saji. Atribut nominal ini memiliki tiga kemungkinan nilai: kecil, sedang, dan besar. Nilai memiliki urutan yang bermakna (yang sesuai dengan peningkatan ukuran minuman); namun, kita tidak dapat mengetahui dari nilai seberapa besar, katakanlah, sebuah media daripada yang besar. Contoh atribut ordinal lainnya termasuk grade (mis., A+, A, A−, B+, dan seterusnya) dan peringkat profesional. Pangkat profesional dapat disebutkan secara berurutan: misalnya, asisten, rekan, dan penuh untuk profesor, dan swasta, kelas satu swasta, spesialis, kopral, dan sersan untuk pangkat tentara.

Atribut ordinal berguna untuk mencatat penilaian subjektif atas kualitas yang tidak dapat diukur secara objektif; sehingga atribut ordinal sering digunakan dalam survei untuk pemeringkatan. Dalam satu survei, peserta diminta menilai seberapa puas mereka sebagai pelanggan. Kepuasan pelanggan memiliki urutan kategori sebagai berikut: 0: sangat tidak puas, 1: agak tidak puas, 2: netral, 3: puas, dan 4: sangat puas.

Atribut ordinal juga dapat diperoleh dari diskritisasi kuantitas numerik dengan membagi rentang nilai menjadi sejumlah kategori terurut yang terbatas seperti yang dijelaskan dalam Bab 3 tentang reduksi data.

*Central tendency* dari atribut ordinal dapat diwakili oleh mode dan mediannya (nilai tengah dalam urutan berurutan), tetapi rata-rata tidak dapat ditentukan.

Perhatikan bahwa atribut nominal, biner, dan ordinal bersifat kualitatif. Artinya, mereka mendeskripsikan fitur suatu objek tanpa memberikan ukuran atau kuantitas sebenarnya. Nilai dari atribut kualitatif tersebut biasanya adalah kata-kata yang mewakili kategori. Jika digunakan bilangan bulat, bilangan bulat mewakili kode komputer untuk kategori tersebut, berlawanan dengan jumlah terukur (misalnya, 0 untuk ukuran minuman kecil, 1 untuk ukuran sedang, dan 2 untuk ukuran besar).

#### Atribut Numerik
Atribut numerik bersifat kuantitatif; yaitu, itu adalah kuantitas yang dapat diukur, direpresentasikan dalam bilangan bulat atau nilai nyata. Atribut numerik dapat berskala interval atau berskala rasio.


##### Atribut Berskala Interval
  Atribut berskala interval diukur pada skala unit berukuran sama. Nilai atribut berskala interval memiliki urutan dan bisa positif, 0, atau negatif. Jadi, selain memberikan peringkat nilai, atribut semacam itu memungkinkan kita untuk membandingkan dan mengukur perbedaan antar nilai.
Contoh atribut berskala Interval. Atribut suhu berskala interval. Misalkan kita memiliki nilai suhu luar ruangan untuk beberapa hari yang berbeda, dimana setiap hari adalah sebuah objek. Dengan mengurutkan nilai, kami memperoleh peringkat objek sehubungan dengan suhu. Selain itu, kita dapat mengukur perbedaan antara nilai-nilai. Misalnya, suhu 20◦C lima derajat lebih tinggi dari suhu 15◦C. Tanggal kalender adalah contoh lain. Misalnya, tahun 2002 dan 2010 berjarak delapan tahun.

Suhu dalam Celcius dan Fahrenheit tidak memiliki titik nol sebenarnya, yaitu, baik 0 ◦C maupun 0◦F tidak menunjukkan “tidak ada suhu”. (Pada skala Celcius, misalnya, satuan pengukurannya adalah 1/100 perbedaan antara suhu leleh dan suhu didih air dalam tekanan atmosfer.) Meskipun kita dapat menghitung perbedaan antara nilai suhu, kita tidak dapat membicarakan satu nilai suhu sebagai kelipatan dari yang lain. Tanpa nol sejati, kita tidak dapat mengatakan, misalnya, bahwa 10◦C dua kali lebih hangat dari 5◦C. Artinya, kita tidak dapat berbicara tentang nilai-nilai dalam bentuk rasio. Demikian pula, tidak ada titik nol sebenarnya untuk tanggal kalender. (Tahun 0 tidak sesuai dengan awal waktu.) Ini membawa kita ke atribut berskala rasio, yang keluar dari titik nol yang sebenarnya.
Karena atribut berskala interval bersifat numerik, kita dapat menghitung nilai rata-ratanya, selain ukuran median dan modus tendensi sentral.

##### Atribut Berskala Rasio
Atribut skala rasio adalah atribut numerik dengan titik nol yang melekat. Artinya, jika pengukuran berskala rasio, kita dapat berbicara tentang suatu nilai sebagai kelipatan (atau rasio) dari nilai lain. Selain itu, nilai-nilai diurutkan, dan kita juga dapat menghitung selisih antara nilai-nilai, serta rata-rata, median, dan modus.

**Contoh atribut berskala rasio :**

Tidak seperti suhu dalam Celsius dan Fahrenheit, skala suhu Kelvin (K) dianggap sebagai titik nol sebenarnya (0◦K = −273,15◦C): Ini adalah titik di mana partikel penyusun materi memiliki energi kinetik nol. Contoh lain dari atribut berskala rasio termasuk menghitung atribut seperti pengalaman bertahun-tahun (misalnya, objek adalah karyawan) dan jumlah kata (misalnya, objek adalah dokumen). Contoh tambahan termasuk atribut untuk mengukur koordinat berat, tinggi, lintang dan bujur (mis., saat mengelompokkan rumah), dan jumlah uang (mis., Anda 100 kali lebih kaya dengan 100 dolar dibandingkan dengan 1 dolar).

##### Atribut Diskrit versus Kontinu
Dalam presentasi kami, kami telah mengatur atribut menjadi tipe nominal, biner, ordinal, dan numerik. Ada banyak cara untuk mengatur tipe atribut. Jenisnya tidak saling eksklusif.

Algoritme klasifikasi yang dikembangkan dari bidang pembelajaran mesin sering berbicara tentang atribut sebagai diskrit atau kontinu. Setiap jenis dapat diproses secara berbeda. Atribut diskrit memiliki kumpulan nilai yang terbatas atau tak terbatas, yang mungkin atau tidak mungkin direpresentasikan sebagai bilangan bulat. Atribut warna rambut, perokok, tes medis, dan ukuran minuman masing-masing memiliki jumlah nilai yang terbatas, dan juga diskrit. Perhatikan bahwa atribut diskrit mungkin memiliki nilai numerik, seperti 0 dan 1 untuk atribut biner atau, nilai 0 hingga 110 untuk usia atribut. Atribut dihitung tak terbatas jika himpunan nilai yang mungkin tak terbatas tetapi nilainya dapat diletakkan dalam korespondensi *one-to-one* dengan bilangan asli. Misalnya, ID pelanggan atribut terhitung tak terbatas. Jumlah pelanggan dapat bertambah hingga tak terhingga, tetapi pada kenyataannya, kumpulan nilai sebenarnya dapat dihitung (di mana nilai dapat diletakkan dalam korespondensi *one-to-one* dengan kumpulan bilangan bulat). Kode pos adalah contoh lain.

Jika suatu atribut tidak diskrit, maka atribut tersebut kontinu. Istilah atribut numerik dan atribut kontinyu sering digunakan secara bergantian dalam literatur. (Hal ini dapat membingungkan karena, dalam pengertian klasik, nilai kontinu adalah bilangan real, sedangkan nilai numerik dapat berupa bilangan bulat atau bilangan real.) Dalam praktiknya, nilai real direpresentasikan menggunakan jumlah digit yang terbatas. Atribut kontinu biasanya direpresentasikan sebagai variabel *floating-point*.

# Deskripsi Statistik Dasar Data

## Mengukur *Central Tendency*: Mean, Median, dan Modus
Pada bagian ini, kita melihat berbagai cara untuk mengukur *Central Tendency* data. Misalkan kita memiliki beberapa atribut X, seperti gaji, yang telah dicatat untuk sekumpulan objek. Misalkan $x_1,x_2,...,x_N$ adalah himpunan $N$ nilai pengamatan atau pengamatan untuk $X$. Di sini, nilai ini juga dapat disebut sebagai kumpulan data (untuk $X$). Jika kita memplot pengamatan untuk gaji, di mana sebagian besar nilainya akan jatuh? Ini memberi kita gambaran tentang *Central Tendency* dari data. Ukuran tendensi sentral meliputi mean, median, modus, dan midrange. Ukuran numerik paling umum dan efektif dari "*Central*" sekumpulan data adalah *Mean*(aritmatika). Misalkan $x_1,x_2,...,x_N$ adalah sekumpulan nilai atau observasi $N$, seperti atribut numerik $X$, seperti gaji. Rata-rata *(Mean)* dari himpunan nilai tersebut adalah

$$
\overline{x}=\frac{\sum \limits_{i=1}^{N}x_i}{N}=\frac{x_1 + x_2+ ... + x_N}{N}\
$$

Ini sesuai dengan fungsi agregat bawaan, rata-rata (*avg()* dalam SQL), yang disediakan dalam *relational database systems*.

Agar prapemrosesan data berhasil, penting untuk memiliki gambaran keseluruhan tentang data Anda. Deskripsi statistik dasar dapat digunakan untuk mengidentifikasi properti data dan menyoroti nilai data mana yang harus diperlakukan sebagai noise atau outlier.
Bagian ini membahas tiga bidang deskripsi statistik dasar. Kita mulai dengan ukuran tendensi sentral (Bagian 2.2.1), yang mengukur lokasi tengah atau pusat distribusi data. Berbicara secara intuitif, mengingat atribut, di mana sebagian besar nilainya jatuh? Secara khusus, kita membahas rata-rata, median, modus, dan midrange.
  Selain menilai kecenderungan sentral dari kumpulan data kami, kami juga ingin memiliki gagasan tentang penyebaran data. Artinya, bagaimana data tersebar? Ukuran dispersi data yang paling umum adalah jangkauan, kuartil, dan jangkauan interkuartil; ringkasan lima angka dan plot kotak; dan varians dan standar deviasi data Langkah-langkah ini berguna untuk mengidentifikasi outlier dan dijelaskan dalam Bagian 2.2.2.
Terakhir, kita dapat menggunakan banyak tampilan grafis dari deskripsi statistik dasar untuk memeriksa data kita secara visual (Bagian 2.2.3). Sebagian besar paket perangkat lunak penyajian data statistik atau grafis mencakup diagram batang, diagram lingkaran, dan grafik garis. Tampilan ringkasan dan distribusi data populer lainnya termasuk plot kuantil, plot kuantil-kuantil, histogram, dan plot pencar.

Mengukur Tendensi Sentral: Mean, Median, dan Modus
Pada bagian ini, kita melihat berbagai cara untuk mengukur tendensi sentral data. Misalkan kita memiliki beberapa atribut X, seperti gaji, yang telah dicatat untuk sekumpulan objek. Misalkan x1,x2,...,xN adalah himpunan N nilai pengamatan atau pengamatan untuk X. Di sini, nilai ini juga dapat disebut sebagai kumpulan data (untuk X). Jika kita memplot pengamatan untuk gaji, di mana sebagian besar nilainya akan jatuh? Ini memberi kita gambaran tentang kecenderungan sentral dari data. Ukuran tendensi sentral meliputi mean, median, modus, dan midrange. Ukuran numerik paling umum dan efektif dari "pusat" sekumpulan data adalah rata-rata (aritmatika). Ini sesuai dengan fungsi agregat bawaan, rata-rata (rata-rata () dalam SQL), yang disediakan dalam sistem basis data relasional.

**Contoh Mean :** Misalkan kita memiliki nilai berikut untuk gaji (dalam ribuan dolar), ditunjukkan dalam urutan meningkat: 30, 36, 47, 50, 52, 52, 56, 60, 63, 70, 70, 110. Maka:

$$
\overline{x}=\frac{30 + 36 + 47 + 50 + 52 + 52 + 56 + 60 + 63 + 70 + 70 + 110}{12}=\frac{696}{12}=58
$$

Jadi, gaji rata-rata adalah $58.000.

Terkadang, setiap nilai $x_i$ dalam suatu himpunan dapat dikaitkan dengan bobot $w_i$ untuk $i = 1,...,N$. Bobot mencerminkan signifikansi, kepentingan, atau frekuensi kejadian yang melekat pada nilainya masing-masing. Dalam hal ini, dapat dihitung dengan menggunakan rumus:

$$
\overline{x}=\frac{\sum \limits_{i=1}^{N}w_ix_i}{\sum \limits_{i=1}^{N}w_i}=\frac{w_1x_1 + w_2x_2+ ... + w_Nx_N}{w_1 + w_2+ ... + w_N}\
$$

Ini disebut ***weighted arithmetic mean*** or ***the weighted average***.
